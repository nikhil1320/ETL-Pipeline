```python
from sqlalchemy import create_engine
import urllib.parse

# MySQL database credentials
username = "root"
password = urllib.parse.quote_plus("root")  # URL-encode the password
host = "127.0.0.1"       # Use "127.0.0.1" or "localhost"
port = "3306"            # MySQL default port
database_name = "ecommerce_db"

# Create the SQLAlchemy engine
engine = create_engine(f"mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database_name}")

# Test the connection
try:
    conn = engine.connect()
    print("Connection successful!")
    conn.close()
except Exception as e:
    print(f"Connection failed: {e}")
```

```python
#Upload dataframe to mysql table
df_c.to_sql('customers',con=engine, if_exists = 'replace', index=False)
df_o.to_sql('orders',con=engine, if_exists = 'replace', index=False)
df_p.to_sql('products',con=engine, if_exists = 'replace', index=False)
df_pay.to_sql('payments',con=engine, if_exists = 'replace', index=False)
```

```python
var = """ select c.customer_id,count(o.order_id) as Total_number_of_Order_Placed
            from Customers c 
            join Orders o 
            on c.customer_id = o.customer_id
            group by 1
            order by 2 desc; """
pd.read_sql(var,conn).head()
```

```python
var = """ select CONCAT(c.First_name," ",c.Last_name) AS Customer_Name ,o.status AS Order_status
FROM Customers AS c 
JOIN Orders AS o
ON c.customer_id = o.customer_id
ORDER BY Customer_Name;"""
pd.read_sql(var,conn).head()
```

```python
var = ' select round(sum(total_amount),2) as Total_Revenue from Orders '
pd.read_sql(var,conn).head()
```

```python
var = """ with cte as (select c.customer_id,concat(c.first_name,' ',c.last_name) as Customer_Name , count(o.order_id) as Total_number_of_Order_Placed
            from Customers c 
            join Orders o 
            on c.customer_id = o.customer_id
            group by 1) 
            
            select customer_id,Customer_Name
            from cte 
            where Total_number_of_Order_Placed > 3 ; """
pd.read_sql(var,conn).head()
```

```python
var = """ select p.payment_status,
            concat(o.order_id,'-',date_format(o.order_date,'%y/%m/%d'),'-',o.total_amount,'-',o.status) as Brif_Order_details
            from Payments p
            join Orders o
            on p.order_id = o.order_id """
pd.read_sql(var,conn).head()
```

```python
var = """ select concat(c.first_name,' ',c.last_name) as Customer_Name , round(sum(o.total_amount),2) as Total_Spent
            from customers c
            join Orders o
            on c.customer_id = o.customer_id
            group by 1 
            order by 2 desc
            limit 5; """
pd.read_sql(var,conn).head()
```

```python
var = """ select * from Payments 
            where payment_status = 'Pending' or payment_status = 'Failed' """
pd.read_sql(var,conn).head()
```

```python
var = """ select p.category , count(o.order_id) as Total_Sales 
            from Products p
            join Orders o
            on p.product_id = o.product_id
            group by p.category
            order by Total_Sales desc; """
pd.read_sql(var,conn).head()
```

```python
var = """ select order_id from Payments
            where payment_status <> 'Completed'; """
pd.read_sql(var,conn).head()
```

```python
var = """ select concat(c.first_name,' ',c.last_name) as Customer_Name , round(avg(o.total_amount),2) as Avg_Order_Value
            from Customers c
            join Orders o
            on c.customer_id = o.customer_id
            group by Customer_Name
            order by Avg_Order_Value desc; """
pd.read_sql(var,conn).head()
```

```python
var = """ select payment_method , count(*) as Total_Used 
            from Payments
            group by 1
            order by 2 desc
            limit 1; """
pd.read_sql(var,conn).head()
```

```python
var = """ SELECT concat(round((SUM(CASE WHEN status = 'Cancelled' THEN 1 ELSE 0 END) / COUNT(*) * 100),2),'%') AS `Cancelled_%`
            FROM Orders; """
pd.read_sql(var,conn).head()
```

```python
var = """ SELECT monthname(order_date) AS 'Month', COUNT(*) AS Number_Of_Orders
            FROM Orders
            GROUP BY monthname(order_date)
            ORDER BY Number_Of_Orders DESC
            LIMIT 1; """
pd.read_sql(var,conn).head()
```
